## Ejecutar una consulta KQL con PySpark
- Referencia: https://learn.microsoft.com/es-es/fabric/real-time-intelligence/notebooks
- Code snippet: Teclear read en una celda

```python
kusto_uri = "https://trd-2ug9gw3yqjx61qz020.z1.kusto.fabric.microsoft.com"
kusto_database = "Casa_Eventos"
kusto_query = "Stocks | take 100"

df  = spark.read\
            .format("com.microsoft.kusto.spark.synapse.datasource")\
            .option("accessToken", mssparkutils.credentials.getToken(kustoUri))\
            .option("kustoCluster", kusto_uri)\
            .option("kustoDatabase", kusto_database) \
            .option("kustoQuery", kusto_query).load()

display(df)
```
