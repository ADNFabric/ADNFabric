### Carga un archivo Parquet hacia una tabla en el Lakehouse BRONCE¨
- Corrige los nombres de las columnas para que sean compatibles con Delta

```python
# Parámetros

RUTA_ARCHIVO_PARQUET = 'Files/AdventureWorks/Production.Product.parquet'
NOMBRE_TABLA = 'Production_Product'
```

```python
# Funciones auxiliares

import re 

def sustituir_letras_acentuadas(texto):
    reemplazos = {
        'á': 'a', 'é': 'e', 'è': 'e', 'í': 'i', 'ó': 'o', 'ú': 'u'
    }
    return ''.join(reemplazos.get(c, c) for c in texto)


def corregir_nombres_columnas(columnas):
    columnas_nuevas = []
    for c in columnas:
        c = sustituir_letras_acentuadas(c)
        c = re.sub('[,;{}()=]','',c)
        c = re.sub('[ \n\t]','_',c)
        columnas_nuevas.append(c)
    return columnas_nuevas
```

```python
# Leyendo los datos hacia un DF Spark
df = spark.read.parquet(RUTA_ARCHIVO_PARQUET)
```

```python
# Corrigiendo los nombres d elas columnas
columnas_corregidas = corregir_nombres_columnas(df.columns)
df = df.toDF(*columnas_corregidas)
```

```python
# Copiando los datos hacia la tabla
df.write.format('delta').mode('overwrite').saveAsTable(NOMBRE_TABLA)
```
