# Transformar las tablas Production_Product, Production_ProductCategory y Production_ProductSubcategory


```python
# Cargar en un dataframe la tabla Productos
df_product = spark.table("Production_Product")
```

```python
df_product = df_product.select(  
    "ProductID",
    "Name",
    "ProductNumber",
    "MakeFlag",
    "FinishedGoodsFlag",
    "Color",
    "SafetyStockLevel",
    "ReorderPoint",
    "StandardCost",
    "ListPrice",
    "Size",
    "SizeUnitMeasureCode",
    "WeightUnitMeasureCode",
    "Weight",
    "DaysToManufacture",
    "ProductLine",
    "Class",
    "Style",
    "ProductSubcategoryID",
    "ProductModelID",
    "SellStartDate",
    "SellEndDate",
    "DiscontinuedDate",
    "ModifiedDate"
)
```

```python
display(df_product.limit(10))
```

```python
df_category = spark.table("Production_ProductCategory")
```

```python
df_category = df_category.select("ProductCategoryID", "Name")
```

```python
df_category = df_category.withColumnRenamed("Name", "Category")
```

```python
display(df_category.limit(10))
```

```python
df_subcategory = spark.table("Production_ProductSubcategory")
```

```python
df_subcategory = df_subcategory.select("ProductSubcategoryID", "Name", "ProductCategoryID")
```

```python
df_subcategory = df_subcategory.withColumnRenamed("Name", "Subcategory")
```

```python
display(df_subcategory.limit(10))
```

```python
df_product = df_product.join(df_subcategory, on="ProductSubcategoryID", how="left") 
```

```python
display(df_product.limit(10))
```

```python
df_product = df_product.join(df_category, on="ProductCategoryID", how="left") 
```

```python
display(df_product.limit(10))
```

```python
df_product = df_product.drop("ProductCategoryID","ProductSubcategoryID")
```

```python
display(df_product.limit(10))
```

```python
#  creará la tabla como una tabla gestionada en el Lakehouse (almacenada en Tables/)
df_product.write.mode("overwrite").format("delta").saveAsTable("LH_STORE_Plata.Production_Product")
```

# Transformar Sales_SalesTerritory

```python
# Cargar la tabla Sales_SalesTerritory en un Dataframe
df_territory = spark.table("Sales_SalesTerritory")
```

```python
# Seleccionar las columnas
df_territory = df_territory.select("TerritoryID", "Name", "CountryRegionCode", "Group", "ModifiedDate")
```

# Transformar las tablas Sales_SalesOrderHeader y Sales_SalesOrderDetail

```python
# Selecciona los campos de la tabla y los caraga en el DataFrame df_header
df_header = spark.table("Sales_SalesOrderHeader")
```

```python
df_header = df_header.select("SalesOrderID", "OrderDate", "DueDate", "ShipDate", "TerritoryID")
```

```python
# Selecciona los campos de la tabla y los caraga en el DataFrame df_detail
df_detail = spark.table("Sales_SalesOrderDetail")
```

```python
df_detail = df_detail.select("SalesOrderID", "OrderQty", "ProductID", "UnitPrice", "UnitPriceDiscount", "ModifiedDate")
```

```python
# Combina los Dataframes df_detail y df_header usando la función join
df_detail = df_detail.join(df_header, on="SalesOrderID", how="left")
display(df_detail.limit(10))
```

```python
#  Crear la tabla Sales_SalesOrderDetail en el Lakehouse LH_STORE_Plata
df_detail.write.mode("overwrite").format("delta").saveAsTable("LH_STORE_Plata.Sales_SalesOrderDetail")
```
