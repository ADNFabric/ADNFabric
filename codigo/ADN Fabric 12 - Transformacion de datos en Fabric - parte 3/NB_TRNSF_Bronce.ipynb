### Carga los archivos Parquet dentro de una carpeta en tablas del lakehouse BRONCE
- Utiliza el cuaderno NB_TRNSF_ParquetATabla
- El nombre de la tabla es el mismo nombre del archivo, sustituyendo el . por _

```python
RUTA_CARPETA = 'Files/AdventureWorks'
```

```python
from pathlib import Path 

for f in notebookutils.fs.ls(RUTA_CARPETA):
    ruta_archivo_parquet = f.path
    nombre_archivo = Path(f.name).stem
    nombre_tabla = nombre_archivo.replace('.','_')
    print(f'Cargando el archivo {nombre_archivo} hacia la tabla {nombre_tabla}')
    notebookutils.notebook.run('NB_TRNSF_ParquetATabla', 120, {'RUTAARCHIVO_PARQUET': ruta_archivo_parquet, 'NOMBRE_TABLA': nombre_tabla})
```
